```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.ion()
%matplotlib widget
```

```python
# Read data

# Only EOM
EOM_data = pd.read_csv('Data/WaveData20240530_007.csv')

# With EDFA on
EDFA_data = pd.read_csv('Data/WaveData20240530_008.csv')
```

```python
# Drop first 26 rows
EOM_data = EOM_data.drop(EOM_data.index[0:26])
EDFA_data = EDFA_data.drop(EDFA_data.index[0:26])


# Drop last column
EOM_data = EOM_data.drop(EOM_data.columns[-1], axis=1)
EDFA_data = EDFA_data.drop(EDFA_data.columns[-1], axis=1)


# Name columns wavelength/voltage
EOM_data.columns = ['Wavelength', 'Voltage']
EDFA_data.columns = ['Wavelength', 'Voltage']
```

```python
wavelengths_EOM = pd.to_numeric(EOM_data['Wavelength'].to_numpy())
voltages_EOM = pd.to_numeric(EOM_data['Voltage'].to_numpy())

wavelengths_EDFA = pd.to_numeric(EDFA_data['Wavelength'].to_numpy())
voltages_EDFA = pd.to_numeric(EDFA_data['Voltage'].to_numpy())
```

```python
power_EOM = 1.2
power_EDFA = 1.88

ratio = power_EDFA/power_EOM

# Plot data in a loglog EOM
plt.figure()
plt.loglog(wavelengths_EOM, voltages_EOM, label='EOM')
plt.loglog(wavelengths_EDFA, voltages_EDFA/ratio, label='EDFA')
plt.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('Voltage (V)')
plt.show()


```
