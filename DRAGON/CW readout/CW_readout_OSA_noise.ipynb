```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as ml

plt.ion()
%matplotlib widget
```

```python
santec_1nm = np.loadtxt(f'Data/OSA noise/WaveData20240802_020.csv', delimiter=',', skiprows=30) # 1nm
santec_50nm = np.loadtxt(f'Data/OSA noise/WaveData20240802_021.csv', delimiter=',', skiprows=30) # 50 nm
EDFA_10nm = np.loadtxt(f'Data/OSA noise/WaveData20240802_022.csv', delimiter=',', skiprows=30) # 10 nm
```

```python
index_max_santec_1nm = np.where(santec_1nm[:,1] == np.max(santec_1nm[:,1]))[0][0]
index_max_santec_50nm = np.where(santec_50nm[:,1] == np.max(santec_50nm[:,1]))[0][0]
index_max_EDFA_10nm = np.where(EDFA_10nm[:,1] == np.max(EDFA_10nm[:,1]))[0][0]
```

```python
wl_santec_1nm = santec_1nm[index_max_santec_1nm, 0]
wl_santec_50nm = santec_50nm[index_max_santec_50nm, 0]
wl_EDFA_10nm = EDFA_10nm[index_max_EDFA_10nm, 0]

print(f'Maximum power for Santec 1nm: {wl_santec_1nm} nm')
print(f'Maximum power for Santec 50nm: {wl_santec_50nm} nm')
print(f'Maximum power for EDFA 10nm: {wl_EDFA_10nm} nm')
```

```python
# Determine a 6GHz window around the peak
f_center = 3e8/(wl_EDFA_10nm*1e-9)

f_lower = f_center - 3e9
f_upper = f_center + 3e9

lambda_lower = 3e8/f_upper
lambda_upper = 3e8/f_lower

print(f'Center frequency: {f_center/1e12} THz')
print(f'Lower frequency: {f_lower/1e12} THz')
print(f'Upper frequency: {f_upper/1e12} THz')
print(f'Lower wavelength: {lambda_lower*1e9} nm')
print(f'Upper wavelength: {lambda_upper*1e9} nm')
print(f'Bandwidth of 6GHz window in nm: {(lambda_upper-lambda_lower)*1e9} nm')

```

```python
# Integrate EDFA from wl_EDFA_10nm-5 to lambda_lower and from lambda_upper to wl_EDFA_10nm+5
index_lower = np.where(EDFA_10nm[:,0] > lambda_lower)[0][0]
index_upper = np.where(EDFA_10nm[:,0] > lambda_upper)[0][0]

power_lower = np.sum(EDFA_10nm[:index_lower, 1])
power_upper = np.sum(EDFA_10nm[index_upper:, 1])

power_in_window = np.sum(EDFA_10nm[index_lower:index_upper, 1])

print(f'Power in 6 GHz window: {power_in_window} dBm')
print(f'Power outside 6 GHz window: {power_lower+power_upper} dBm')

ratio = power_in_window/(power_lower+power_upper)
print(f'Ratio of power in 6 GHz window to power outside 6 GHz window: {ratio}')
```
