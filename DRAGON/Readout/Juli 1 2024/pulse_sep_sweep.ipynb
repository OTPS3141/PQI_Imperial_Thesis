```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pulse_sep = np.array([1, 1.2, 1.4, 1.5, 1.6, 1.8, 2])
readout_eff = ([30.5, 24.7, 16.75, 13.86, 8.5, 6, 4.23])

plt.plot(pulse_sep, readout_eff, 'o-', label='Readout efficiency')
plt.xlabel('Storage time (ns)')
plt.ylabel('Readout efficiency (%)')
plt.title('Readout efficiency vs storage time')
plt.legend()
plt.grid()
plt.show()
```
