```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_joule_per_pulse(EOM_min, avg_power_w_pulse, pulse_sep):
    avg_power_from_pulse = (avg_power_w_pulse - EOM_min)/1000 
    num_pulses_per_second = 1/(pulse_sep*1e-9)
    return (avg_power_from_pulse/num_pulses_per_second)*1e12
```

```python
data_5_ns = np.array([
    [100, 0.78, 5.7750, 1.78, 5.7750, 5.5625],
    [200, 2.45, 5.7000, 5.25, 5.7000, 5.2500],
    [300, 4.24, 5.5500, 8.70, 5.6625, 5.0125],
    [400, 6.25, 5.4250, 12.60, 5.6250, 4.7250],
    [500, 7.80, 5.1125, 15.90, 5.6375, 4.6375]
])

max_5_ns = 5.8
min_5_ns = 0.56
```

```python
data_20_ns = np.array([
    [100, 0.75, 5.7250, 1.54, 5.7125, 5.1250],
    [200, 2.50, 5.5500, 4.80, 5.6625, 4.3125],
    [300, 4.40, 5.3875, 8.07, 5.6375, 3.8875],
    [400, 6.10, 5.2375, 11.30, 5.5625, 3.6750],
    [500, 7.60, 5.1000, 14.37, 5.5875, 3.6375]
])

max_20_ns = 5.8
min_20_ns = 0.56
```

```python
data_50_ns = np.array([
    [100, 0.750, 6.5125, 1.65, 6.375, 5.4750],
    [200, 2.640, 6.4125, 4.28, 6.600, 4.2375],
    [300, 4.30, 6.2500, 7.50, 6.5750, 4.1500],
    [400, 6.10, 6.1375, 10.81, 6.5875, 4.1000],
    [500, 7.85, 5.9375, 13.35, 6.5125, 4.0375]
])

max_50_ns = 6.57
min_50_ns = 0.525
```

```python
data_100_ns = np.array([
    [100, 0.84, 6.7250, 1.31, 6.7750, 4.8875],
    [200, 2.73, 6.5000, 4.19, 6.6750, 4.0500],
    [300, 4.98, 6.1625, 7.15, 6.5875, 4.0375],
    [400, 7.02, 5.7875, 10.03, 6.5250, 3.9500],
    [500, 8.75, 5.6625, 12.55, 6.3500, 4.0125]
])

max_100_ns = 6.8
min_100_ns = 0.55
```

```python
# Assign names to columns: 'EDFA current', 'noAWGpower', 'noAWGabs', 'AWGpower', 'AWGabs'

# data_1_ns to DataFrame
df_1_ns = pd.DataFrame(data_1_ns, columns=['EDFA current', 'noAWGpower', 'noAWGabs', 'AWGpower', 'AWGabsback', 'AWGabs'])

# data_5_ns to DataFrame
df_5_ns = pd.DataFrame(data_5_ns, columns=['EDFA current', 'noAWGpower', 'noAWGabs', 'AWGpower', 'AWGabsback', 'AWGabs'])

# data_20_ns to DataFrame
df_20_ns = pd.DataFrame(data_20_ns, columns=['EDFA current', 'noAWGpower', 'noAWGabs', 'AWGpower', 'AWGabsback', 'AWGabs'])

# data_50_ns to DataFrame
df_50_ns = pd.DataFrame(data_50_ns, columns=['EDFA current', 'noAWGpower', 'noAWGabs', 'AWGpower', 'AWGabsback', 'AWGabs'])

# data_100_ns to DataFrame
df_100_ns = pd.DataFrame(data_100_ns, columns=['EDFA current', 'noAWGpower', 'noAWGabs', 'AWGpower', 'AWGabsback', 'AWGabs'])
```

```python
# Add column with max value

df_50_ns['max'] = np.array([max_50_ns, 6.7, 6.7, 6.7, 6.7])


# Add column to df with absorption values
df_1_ns['absorptionNoAWG'] = (max_1_ns - df_1_ns['noAWGabs'])/max_1_ns
df_1_ns['absorptionDipsAWG'] = (df_1_ns['AWGabsback'] - df_1_ns['AWGabs'])/df_1_ns['AWGabsback']
df_1_ns['absorptionCWAWG'] = (max_1_ns - df_1_ns['AWGabsback'])/max_1_ns
df_1_ns['absorptionDipsTotAWG'] = (max_1_ns - df_1_ns['AWGabs'])/max_1_ns

df_5_ns['absorptionNoAWG'] = (max_5_ns - df_5_ns['noAWGabs'])/max_5_ns
df_5_ns['absorptionDipsAWG'] = (df_5_ns['AWGabsback'] - df_5_ns['AWGabs'])/df_5_ns['AWGabsback']
df_5_ns['absorptionCWAWG'] = (max_5_ns - df_5_ns['AWGabsback'])/max_5_ns
df_5_ns['absorptionDipsTotAWG'] = (max_5_ns - df_5_ns['AWGabs'])/max_5_ns

df_20_ns['absorptionNoAWG'] = (max_20_ns - df_20_ns['noAWGabs'])/max_20_ns
df_20_ns['absorptionDipsAWG'] = (df_20_ns['AWGabsback'] - df_20_ns['AWGabs'])/df_20_ns['AWGabsback']
df_20_ns['absorptionCWAWG'] = (max_20_ns - df_20_ns['AWGabsback'])/max_20_ns
df_20_ns['absorptionDipsTotAWG'] = (max_20_ns - df_20_ns['AWGabs'])/max_20_ns

df_50_ns['absorptionNoAWG'] = (df_50_ns['max'] - df_50_ns['noAWGabs'])/df_50_ns['max']
df_50_ns['absorptionDipsAWG'] = (df_50_ns['AWGabsback'] - df_50_ns['AWGabs'])/df_50_ns['AWGabsback']
df_50_ns['absorptionCWAWG'] = (df_50_ns['max'] - df_50_ns['AWGabsback'])/df_50_ns['max']
df_50_ns['absorptionDipsTotAWG'] = (df_50_ns['max'] - df_50_ns['AWGabs'])/df_50_ns['max']

df_100_ns['absorptionNoAWG'] = (max_100_ns - df_100_ns['noAWGabs'])/max_100_ns
df_100_ns['absorptionDipsAWG'] = (df_100_ns['AWGabsback'] - df_100_ns['AWGabs'])/df_100_ns['AWGabsback']
df_100_ns['absorptionCWAWG'] = (max_100_ns - df_100_ns['AWGabsback'])/max_100_ns
df_100_ns['absorptionDipsTotAWG'] = (max_100_ns - df_100_ns['AWGabs'])/max_100_ns




```

```python
pulse_sep = np.array([1, 5, 20, 50, 100])
EDFA_current = np.array([100, 200, 300, 400, 500])

# Matrix with absorption values for each pulse separation and EDFA current

heatmap_absNoAWG = np.zeros((5, 5))
heatmap_absDipsAWG = np.zeros((5, 5))
heatmap_absCWAWG = np.zeros((5, 5))
heatmap_absDipsTotAWG = np.zeros((5, 5))

heatmap_pulse_power = np.zeros((5, 5))


# fill the matrix with the absorption values

for i in range(5):
    heatmap_absNoAWG[i, 0] = df_1_ns['absorptionNoAWG'][i]
    heatmap_absNoAWG[i, 1] = df_5_ns['absorptionNoAWG'][i]
    heatmap_absNoAWG[i, 2] = df_20_ns['absorptionNoAWG'][i]
    heatmap_absNoAWG[i, 3] = df_50_ns['absorptionNoAWG'][i]
    heatmap_absNoAWG[i, 4] = df_100_ns['absorptionNoAWG'][i]

    heatmap_absDipsAWG[i, 0] = df_1_ns['absorptionDipsAWG'][i]
    heatmap_absDipsAWG[i, 1] = df_5_ns['absorptionDipsAWG'][i]
    heatmap_absDipsAWG[i, 2] = df_20_ns['absorptionDipsAWG'][i]
    heatmap_absDipsAWG[i, 3] = df_50_ns['absorptionDipsAWG'][i]
    heatmap_absDipsAWG[i, 4] = df_100_ns['absorptionDipsAWG'][i]

    heatmap_absCWAWG[i, 0] = df_1_ns['absorptionCWAWG'][i]
    heatmap_absCWAWG[i, 1] = df_5_ns['absorptionCWAWG'][i]
    heatmap_absCWAWG[i, 2] = df_20_ns['absorptionCWAWG'][i]
    heatmap_absCWAWG[i, 3] = df_50_ns['absorptionCWAWG'][i]
    heatmap_absCWAWG[i, 4] = df_100_ns['absorptionCWAWG'][i]

    heatmap_absDipsTotAWG[i, 0] = df_1_ns['absorptionDipsTotAWG'][i]
    heatmap_absDipsTotAWG[i, 1] = df_5_ns['absorptionDipsTotAWG'][i]
    heatmap_absDipsTotAWG[i, 2] = df_20_ns['absorptionDipsTotAWG'][i]
    heatmap_absDipsTotAWG[i, 3] = df_50_ns['absorptionDipsTotAWG'][i]
    heatmap_absDipsTotAWG[i, 4] = df_100_ns['absorptionDipsTotAWG'][i]

    heatmap_pulse_power[i, 0] = get_joule_per_pulse(df_1_ns['noAWGpower'][i], df_1_ns['AWGpower'][i], pulse_sep[0])
    heatmap_pulse_power[i, 1] = get_joule_per_pulse(df_5_ns['noAWGpower'][i], df_5_ns['AWGpower'][i], pulse_sep[1])
    heatmap_pulse_power[i, 2] = get_joule_per_pulse(df_20_ns['noAWGpower'][i], df_20_ns['AWGpower'][i], pulse_sep[2])
    heatmap_pulse_power[i, 3] = get_joule_per_pulse(df_50_ns['noAWGpower'][i], df_50_ns['AWGpower'][i], pulse_sep[3])
    heatmap_pulse_power[i, 4] = get_joule_per_pulse(df_100_ns['noAWGpower'][i], df_100_ns['AWGpower'][i], pulse_sep[4])
    


```

```python
# Plot heatmaps next to each other

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Heatmap for no AWG
im = axs[0, 0].imshow(heatmap_absNoAWG)
axs[0, 0].set_xticks(np.arange(5))
axs[0, 0].set_yticks(np.arange(5))
axs[0, 0].set_xticklabels(pulse_sep)
axs[0, 0].set_yticklabels(EDFA_current)
axs[0, 0].set_title('No AWG')
axs[0, 0].set_xlabel('Pulse separation (ns)')
axs[0, 0].set_ylabel('EDFA current (mA)')
# add colorbar
axs[0, 0].figure.colorbar(im, ax=axs[0, 0])
plt.setp(axs[0, 0].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Heatmap for dips AWG
im = axs[0, 1].imshow(heatmap_absDipsAWG)
axs[0, 1].set_xticks(np.arange(5))
axs[0, 1].set_yticks(np.arange(5))
axs[0, 1].set_xticklabels(pulse_sep)
axs[0, 1].set_yticklabels(EDFA_current)
axs[0, 1].set_title('Dips AWG')
axs[0, 1].set_xlabel('Pulse separation (ns)')
axs[0, 1].set_ylabel('EDFA current (mA)')
# add colorbar
axs[0, 1].figure.colorbar(im, ax=axs[0, 1])
plt.setp(axs[0, 1].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Heatmap for CW AWG
im = axs[1, 0].imshow(heatmap_absCWAWG)
axs[1, 0].set_xticks(np.arange(5))
axs[1, 0].set_yticks(np.arange(5))
axs[1, 0].set_xticklabels(pulse_sep)
axs[1, 0].set_yticklabels(EDFA_current)
axs[1, 0].set_title('CW AWG')
axs[1, 0].set_xlabel('Pulse separation (ns)')
axs[1, 0].set_ylabel('EDFA current (mA)')
# add colorbar
axs[1, 0].figure.colorbar(im, ax=axs[1, 0])
plt.setp(axs[1, 0].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Heatmap for dips total AWG
im = axs[1, 1].imshow(heatmap_absDipsTotAWG)
axs[1, 1].set_xticks(np.arange(5))
axs[1, 1].set_yticks(np.arange(5))
axs[1, 1].set_xticklabels(pulse_sep)
axs[1, 1].set_yticklabels(EDFA_current)
axs[1, 1].set_title('Dips Total AWG')
axs[1, 1].set_xlabel('Pulse separation (ns)')
axs[1, 1].set_ylabel('EDFA current (mA)')
# add colorbar
axs[1, 1].figure.colorbar(im, ax=axs[1, 1])
plt.setp(axs[1, 1].get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")


fig.tight_layout()
plt.show()

```

```python
# Plot heatmap for pulse power

plt.matshow(heatmap_pulse_power)
plt.xticks(np.arange(5), pulse_sep)
plt.yticks(np.arange(5), EDFA_current)
plt.xlabel('Pulse separation (ns)')
plt.ylabel('EDFA current (mA)')
plt.title('Pulse power (pJ)')
plt.colorbar()
plt.show()
```
