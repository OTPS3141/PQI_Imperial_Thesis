## Get mapping popShift to fit distance

```python
def get_elecsus(freq_array, Rb_spectrum, popShift87, T):

    # Fit ElecSus

    import ElecSus_PQI.elecsus.elecsus_methods as EM

    # Horizontal polarization
    E_in=np.array([0,1,0])
    E_in_angle = [E_in[0].real,[abs(E_in[1]),np.angle(E_in[1])]]
    E_in=E_in_angle

    # Cicular polarisation
    # phase = np.pi
    # E_in=np.array([1,1,0])
    # E_in = [E_in[0], E_in[1]*np.exp(1.j*phase)]
    # E_in_angle = [E_in[0].real,[abs(E_in[1]),np.angle(E_in[1])]]
    # E_in = E_in_angle

    groundPop = 0 # 0 for popShift, 1-7 for different F states. So far this is only hardcoded for Rb87

    # Add Gaussian for insufficient linewidth pump
    add_gaussian = False

    # popShift denotes the shift in population. If positive, lowest F state population is increased, if negative, lowest F state population is decreased


    # Initial values
    # T = 72.5
    delta_T = 0.9 * T
    p_dict = {'Elem':'Rb','Dline':'D2','T':T,'lcell':130e-3, 'shift':0, 'Btheta':0,'Bfield':0,
                'shift':0,'Bphi':0,'GammaBuf':0, 'DoppTemp': T, 'rb85frac': 2, 'Constrain':True, 'popShift87': popShift87, 'popShift85': 0, 'K40frac': 0, 'K41frac': 0,
                'sigma': 0.3, 'X0_shift': 0, 'gaussianPop': 1.5, 'GammaBuf': 0}
    p_dict_guess=p_dict

    p_dict_bools={'T':False, 'DoppTemp':False, 'rb85frac': True, 'shift':True, 'popShift87':False, 'popShift85':True, 'GammaBuf':False, 'gaussianPop':add_gaussian, 
                'sigma': add_gaussian, 'X0_shift': add_gaussian}

    p_dict_bounds={}
    p_dict_bounds['shift']=(-50, 50)
    p_dict_bounds['rb85frac']=(0, 10)
    p_dict_bounds['T']=(T-delta_T, T+delta_T)
    p_dict_bounds['popShift87']=(-5,5)
    p_dict_bounds['popShift85']=(-5,5)
    p_dict_bounds['GammaBuf']=(0, 10)
    p_dict_bounds['gaussianPop']=(0, 5)
    p_dict_bounds['sigma']=(0.1, 1)
    p_dict_bounds['X0_shift']=(-1,1)

    # p_dict_bounds['DoppTemp']=(0, 40)


    # Fit
    best_params, RMS, result = EM.fit_data([freq_array*1e3, Rb_spectrum], p_dict_guess, p_dict_bools, groundPop, add_gaussian,
                                        E_in, p_dict_bounds, data_type='S0',fit_algorithm='ML', verbose=False)

    # report = result.fit_report()
    elecsus_fit = result.best_fit
    # print(report)

    return elecsus_fit
```

## Fit pumping shape to Voigt-Gaussian convolution

```python
from scipy.special import wofz


def voigt(x, mu, sigma, gamma):
    return np.real(wofz(((x - mu) + 1j*gamma)/(np.sqrt(2)*sigma))) / (sigma*np.sqrt(2*np.pi))

def gaussian(x, mu, sigma, offset):
    return np.exp(-0.5*((x - mu)/sigma)**2) / (sigma*np.sqrt(2*np.pi)) + offset

def pump_shape(x, sigma_voigt, gamma_voigt, mu_voigt, mu_gaussian, sigma_gaussian, mag_gaussian, offset):
    voigt_profile = -voigt(x, mu_voigt, sigma_voigt, gamma_voigt) 
    gaussian_profile =  mag_gaussian * gaussian(x, mu_gaussian, sigma_gaussian, offset)
    return 1+ voigt_profile + gaussian_profile
```

### Convert freq. to velocity using Doppler shift

```python
dist_from_peak_list = []

for freq in MB_region_freq:
    dist_from_peak = freq - MB_region_freq[np.where(efficiency_MB == peak_efficiency)][0]
    dist_from_peak_list.append(dist_from_peak*1e9)

# Convert distance to velocity

f0 = 384.2304*10**12 + mu*10**9


def doppler_vel(f, f0):
    c = 3*10**8
    return (f - f0)*c/f0

vel_list = []


for delta_freq in dist_from_peak_list:
    vel = np.round(doppler_vel(f0+delta_freq, f0))
    vel_list.append(int(vel))

zero_vel_index = np.where(np.array(vel_list) == 0)[0][0]
vel_list_left = vel_list[:zero_vel_index]
vel_list_right = vel_list[zero_vel_index:]
```

### Convert estimated velocity to number density using MB

```python
# Maxwell Boltzmann distribution
def MB_dist(v, T):
    k = 1.38 * 10**-23
    m = 1.44 * 10**-25
    return 4*np.pi * (m/(2*np.pi*k*T))**(3/2) * v**2 * np.exp(-m*v**2/(2*k*T))

v = np.linspace(0, 500, 500)
MB_distribution = MB_dist(v, T)

# Define the number of velocity classes
num_classes = 500

# Calculate the bin edges for the histogram
bin_edges = np.linspace(min(v), max(v), num_classes + 1)

v_hist = np.histogram(v, bins=bin_edges, weights=MB_distribution, density=True)

# Create the histogram
plt.figure()
plt.hist(v, bins=bin_edges, weights=MB_distribution, density=True)
plt.plot(v, MB_distribution)
plt.xlabel('Velocity (m/s)')
plt.ylabel('Probability')
plt.title('Histogram of Velocity Classes')
plt.show()

# List with mean of each bin
mean_list = []
for i in range(len(bin_edges)-1):
    mean = (bin_edges[i] + bin_edges[i+1])/2
    mean_list.append(mean)

theta_range = np.linspace(0, 2*np.pi, 50)

def beam_velocity(theta, v):
    return v*np.cos(theta)

velocity_beam_component_list = []
velocity_prob_list = []

for v, v_prob in zip(mean_list, v_hist[0]):
    for theta in theta_range:
        velocity = beam_velocity(theta, v)
        prob = v_prob
        velocity_beam_component_list.append(velocity)
        velocity_prob_list.append(prob)


```

```python
# Get probability density of pumping efficiency
efficiency_prob = efficiency_ROI * doppler_prob

# Normalise
efficiency_prob = efficiency_prob/np.sum(efficiency_prob)

velocity_matrix = []
efficiency_matrix = []

min_pump_array = np.linspace(0, 1, 1000)

for lower_prob in min_pump_array:
    # Get velocities with more than lower_prob efficiency
    velocity_list = []
    efficiency_prob_list = []
    for vel, eff, prob in zip(vel_list, efficiency_ROI, efficiency_prob):
        if eff > lower_prob:
            velocity_list.append(vel)
            efficiency_prob_list.append(prob)
        
    velocity_matrix.append(velocity_list)
    efficiency_matrix.append(np.sum(efficiency_prob_list))

efficiency_matrix = np.array(efficiency_matrix)
sub_unity_index = np.where(efficiency_matrix < 1)[0][0]

# Plot efficiency_matrix vs. min_pump_array
plt.figure()
plt.plot(min_pump_array, efficiency_matrix, label='Cumulative pumping efficiency')
plt.xlabel('Minimum Pumping Efficiency', fontsize=15, fontfamily="Times New Roman")
plt.ylabel('Number Density', fontsize=15, fontfamily="Times New Roman")
# plt.title(f'Cumulative pumping eff. T={T}C, I={I}mW \n Mean Pump. Efficiency: {np.round(np.sum(efficiency_ROI*efficiency_prob), 3)}, F(0.9): {np.round(efficiency_matrix[np.where(min_pump_array <= 0.9)][-1], 3)}, F(0.99): {np.round(efficiency_matrix[np.where(min_pump_array <= 0.99)][-1], 3)}')
plt.axvline(0.99, color='r', linestyle='--')
plt.axhline(efficiency_matrix[np.where(min_pump_array <= 0.99)][-1], color='r', linestyle='--', label='99% efficiency')
# plt.xlim(min_pump_array[sub_unity_index], 1)
plt.xlim(0.9, 1)
plt.legend(loc='lower left', fontsize=12)
# plt.annotate(f'Considered fraction of total atoms {np.round(prob_pump_region,3)}', (0.905, 0.1), fontsize=12)
# plt.savefig(path_to_figures + 'CW_vel_dist_cumulative.pdf', format='pdf')
plt.show()


# Save image 
# plt.savefig(path_to_data + f'I{I}_' + f'T{T}' + '.png')
# os.chdir('/home/otps3141/Documents/Dokumente/ETH QE/Master Thesis Imperial/Thesis/Code/ElecSus_fitting/Data/Pumping/Pulsed Pumping/AdjustedBeamWidth')
# os.system(f"pdftoppm {I}_{T}.pdf {I}_{T} -png")
# os.chdir(path)


print('Probability mass of atoms with more than 99% pumping efficiency: ', efficiency_matrix[np.where(min_pump_array <= 0.99)][-1])

print('Probability mass of atoms with more than 90% pumping efficiency: ', efficiency_matrix[np.where(min_pump_array <= 0.9)][-1])

print('Mean pumping efficiency: ', np.sum(efficiency_ROI*efficiency_prob))


print(f'Considered fraction of total atoms {np.round(prob_pump_region,3)}')
```
